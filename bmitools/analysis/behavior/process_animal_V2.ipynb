```python

def process_animal_id2(animal_id,
                      root_dir):
    
    import os

    sess = ProcessSession(root_dir,
                          animal_id)

    #
    for session_id in sess.session_ids[1:]:

        #
        sess.recompute_session = False

        #
        sess.session_id = session_id

        #
        sess.verbose=False

        #
        sess.save_dir = os.path.join(sess.root_dir,
                                        sess.animal_id,
                                        sess.session_id,
                                        'results')
        
        # make dir if it does not exist
        if not os.path.exists(sess.save_dir):
            os.makedirs(sess.save_dir)

        #
        sess.skip_dict= True
        sess.load_data()

        # this is not so interesting; may need to do more on this
        sess.show_plots = False

        #
        sess.process_session_traces()

        #
        sess.compute_ensemble_correlations()

        #
        sess.window_size = 30   # seconds
        sess.compute_correlograms_reward_vs_licking()

        #
        sess.min_width_event_onphase = 30
        sess.min_width_event_upphase = 5
        sess.recompute_binarization = False

        #
        sess.std_upphase = 1.0
        sess.percentile_threshold = 0.9999999
        sess.use_upphase = True
        sess.binarize_ensembles()

        #
        #if True:
        sess.window = 30   # in seconds
        sess.recompute_session = True
        print ("... starting upphase correlograms")
        sess.compute_correlograms_ensembles_upphase2()

        #
        sess.window = 30   # in seconds
        print ("... starting fluorescence correlograms") 
        sess.compute_correlograms_ensembles_fluorescence()

        #
        sess.bin_width = 5 # in minutes
        sess.compute_intra_session_reward_histogram()

        #
        sess.show_plots=False
        sess.bin_width_mins = 5 # in minutes
        sess.compute_intra_session_reward_histogram_hist_per_minute()

        #
        sess.bin_width = 5 # in minutes
        sess.compute_intra_session_cell_burst_histogram()

        #
        sess.compute_intra_session_cell_burst_histogram_v2()

        #
        sess.isi_width = 10
        sess.isi_bin_width = 0.250
        sess.compute_intra_session_inter_burst_interval()

    print (" ... done with animal_id: ", animal_id)
    print ('')
```
